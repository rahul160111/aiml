```python
import matplotlib.pyplot as plt 
import numpy as np 

xpoint = np.array([0,6])
ypoint = np.array([0,250])

plt.plot(xpoint,ypoint,'x')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

xpoint = np.array([1,2,6,10])
ypoint = np.array([3,8,1,10])

plt.plot(xpoint,ypoint,'o')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

ypoint = np.array([3,10,5,8,2,6])

plt.plot(ypoint,marker='D')
plt.show()
```
